# Variable	Definition	Key
* survival:	Survival	0 = No, 1 = Yes
* pclass :  Ticket class	1 = 1st, 2 = 2nd, 3 = 3rd
* sex :	Sex	
* Age :	Age in years	
* sibsp :	# of siblings / spouses aboard the Titanic	
* parch	: # of parents / children aboard the Titanic	
* ticket : Ticket number	
* fare : Passenger fare	
* cabin	: Cabin number	
* embarked	: Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton

```python
train2['Age']=train2['Age'].astype('int')
```

```python
train2.head()
```

```python
corrMatt=train2[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare"]]
corrMatt=corrMatt.corr()
print(corrMatt)

mask=np.array(corrMatt)
mask[np.tril_indices_from(mask)]=False
```

```python
fig, ax=plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, mask=mask, vmax=.8, square=True, annot=True)
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(ncols=3)
fig.set_size_inches(12,5)
sns.regplot(x="Age",y="Survived",data=train2,ax=ax1)
sns.regplot(x="SibSp",y="Survived",data=train2,ax=ax2)
sns.regplot(x="Parch",y="Survived",data=train2,ax=ax3)
```
